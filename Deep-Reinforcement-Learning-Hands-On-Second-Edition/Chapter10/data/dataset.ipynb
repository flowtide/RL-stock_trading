테스트 데이터 생성(220101 ~ 240930)

```python
import yfinance as yf
import pandas as pd
import datetime

# 데이터 가져오기
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2024, 9, 30)
kb = yf.download("078930.KS", start=start, end=end)

# MultiIndex를 제거하고 단일 인덱스로 변환한 후 열 이름을 설정
if isinstance(kb.columns, pd.MultiIndex):
    kb.columns = kb.columns.droplevel(0)  # 첫 번째 레벨 제거

# 열 이름을 수동으로 설정
kb.columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

# 데이터 포맷에 맞게 변환
formatted_data = []

for date, row in kb.iterrows():
    # 날짜 포맷
    date_str = date.strftime('%Y%m%d')
    # 시간 생성 (100100, 100200, ..., 100700 등)
    for i in range(1, 8):
        time_str = f"100{i}00"
        formatted_data.append({
            "DATE": date_str,
            "TIME": time_str,
            "OPEN": f"{row['Open']:.7f}",
            "HIGH": f"{row['High']:.7f}",
            "LOW": f"{row['Low']:.7f}",
            "CLOSE": f"{row['Close']:.7f}",
            "VOL": int(row['Volume'])
        })
        
# 데이터프레임으로 변환
formatted_df = pd.DataFrame(formatted_data, columns=["DATE", "TIME", "OPEN", "HIGH", "LOW", "CLOSE", "VOL"])

# CSV 파일로 저장
formatted_df.to_csv("kb_220101_240930.csv", index=False)
```
